```python
import pandas as pd

import datetime

import numpy as np

import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

!pip install yfinance

import yfinance as yf

%matplotlib inline
```

```python
start = "2012-01-01"
end = '2022-06-28'
tcs = yf.download('TCS',start,end)
infy = yf.download('INFY',start,end)
wipro = yf.download('WIPRO.NS',start,end)
appl=yf.download('AAPL',start,end)
google=yf.download('GOOG', start, end)
amazon= yf.download('AMZN', start, end)
tesla= yf.download('TSLA', start, end)
meta= yf.download('META', start, end)
netflix= yf.download('NFLX', start, end)
```

```python
tcs['Open'].plot(label = 'TCS', figsize = (15,7))
infy['Open'].plot(label = "Infosys")
wipro['Open'].plot(label = 'Wipro')
appl['Open'].plot(label="Apple")
google['Open'].plot(label="Google")
amazon['Open'].plot(label="Amazon")
tesla['Open'].plot(label="Tesla")
meta['Open'].plot(label="Meta")
netflix['Open'].plot(label="Netflix")
plt.legend()
plt.title('Stock Prices of TCS, Infosys, Wipro, Apple, Google, Amazon, Tesla, Meta & Netflix')
```

```python
plt.plot(tcs.index, tcs['Adj Close'])
# plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
#Subplots
plt.figure(figsize=(20,20), dpi=200)
f, ax = plt.subplots(3, 3, figsize=(20,20), sharex=True)
# f.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# f.gca().xaxis.set_major_locator(mdates.YearLocator())

ax[0,0].plot(tcs.index, tcs['Adj Close'], color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('TCS');

ax[0,1].plot(infy.index, infy['Adj Close'], color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('INFOSYS');

ax[1,0].plot(wipro.index, wipro['Adj Close'], color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('WIPRO');

ax[1,1].plot(google.index, google['Adj Close'], color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('Google');

ax[0,2].plot(appl.index, appl['Adj Close'], color='orange')
ax[0,2].grid(True)
ax[0,2].tick_params(labelrotation=90)
ax[0,2].set_title('APPLE');

ax[1,2].plot(tesla.index, tesla['Adj Close'], color='m')
ax[1,2].grid(True)
ax[1,2].tick_params(labelrotation=90)
ax[1,2].set_title('TESLA');

ax[2,2].plot(meta.index, meta['Adj Close'], color='c')
ax[2,2].grid(True)
ax[2,2].tick_params(labelrotation=90)
ax[2,2].set_title('META');

ax[2,0].plot(amazon.index, amazon['Adj Close'], color='g')
ax[2,0].grid(True)
ax[2,0].tick_params(labelrotation=90)
ax[2,0].set_title('AMAZON');

ax[2,1].plot(netflix.index, netflix['Adj Close'], color='r')
ax[2,1].grid(True)
ax[2,1].tick_params(labelrotation=90)
ax[2,1].set_title('NETFLIX');
```

```python
#Market Capitalisation
tcs['MarktCap'] = tcs['Open'] * tcs['Volume']
infy['MarktCap'] = infy['Open'] * infy['Volume']
wipro['MarktCap'] = wipro['Open'] * wipro['Volume']
tcs['MarktCap'].plot(label = 'TCS', figsize = (15,7))
infy['MarktCap'].plot(label = 'Infosys')
wipro['MarktCap'].plot(label = 'Wipro')
plt.title('Market Cap')
plt.legend()
```

```python
#Market Capitalisation
meta['MarktCap'] = meta['Open'] * meta['Volume']
amazon['MarktCap'] = amazon['Open'] * amazon['Volume']
appl['MarktCap'] = appl['Open'] * appl['Volume']
netflix['MarktCap'] = netflix['Open'] * netflix['Volume']
google['MarktCap'] = google['Open'] * google['Volume']
meta['MarktCap'].plot(label = 'META', figsize = (15,7))
amazon['MarktCap'].plot(label = 'Amazon')
appl['MarktCap'].plot(label = 'APPLE')
netflix['MarktCap'].plot(label = 'NETFLIX')
google['MarktCap'].plot(label = 'GOOGLE')
plt.title('Market Cap')
plt.legend()
```

```python
data = pd.concat([tcs['Open'],infy['Open'],wipro['Open']],axis = 1)
data.columns = ['TCSOpen','InfosysOpen','WiproOpen']
scatter_matrix(data, figsize = (8,8), hist_kwds= {'bins':250})
plt.show()
```

```python
#Volatility
plt.figure(dpi=120)
tcs['returns'] = (tcs['Close']/tcs['Close'].shift(1)) -1
infy['returns'] = (infy['Close']/infy['Close'].shift(1))-1
wipro['returns'] = (wipro['Close']/wipro['Close'].shift(1)) - 1
tcs['returns'].hist(bins = 100, label = 'TCS', alpha = 0.5, figsize = (15,7))
infy['returns'].hist(bins = 100, label = 'Infosysy', alpha = 0.5)
wipro['returns'].hist(bins = 100, label = 'Wipro', alpha = 0.5)
plt.legend()
```

```python
tcs_adj_close_px = tcs['Adj Close']
# Short moving window rolling mean
tcs['42'] = tcs_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
tcs['252'] = tcs_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
tcs[['Adj Close', '42', '252']].plot(title="TCS")

# Show plot
plt.show()

tesla_adj_close_px = tesla['Adj Close']
# Short moving window rolling mean
tesla['42'] = tesla_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
tesla['252'] = tesla_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
tesla[['Adj Close', '42', '252']].plot(title="TESLA")

# Show plot
plt.show()
```
